```python
import pandas as pd
from os.path import join as pjoin
from gears import PertData, GEARS
```

## Config

```python
data_dir = '../../dataset/'
data_name = 'norman-1000'
split = 'simulation'
result_dir = './results'
seed = 3
epochs = 15
batch_size=10
accumulation_steps=5
test_batch_size=32
hidden_size=512
train_gene_set_size=0.75
highres=0 
lr=0.005 
device='cuda:2'
singlecell_model_path=None # not used
```

## Loading Dataset and Model

```python
pert_data = PertData(data_dir)
pert_data.load(data_path = pjoin(data_dir, data_name))
pert_data.prepare_split(split=split, seed=seed, train_gene_set_size=train_gene_set_size)
pert_data.get_dataloader(batch_size = batch_size, test_batch_size = test_batch_size)
gears_model = GEARS(pert_data, device = device)
```

## Training

```python
gears_model.model_initialize(hidden_size = hidden_size, 
                                model_type = 'emb',
                                load_path=singlecell_model_path,
                                finetune_method='frozen',
                                accumulation_steps=accumulation_steps,
                                highres=highres)
gears_model.train(epochs = epochs, result_dir=result_dir,lr=lr)
# gears_model.save_model(result_dir)
```
